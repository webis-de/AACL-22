```python
from utils import *
import pandas as pd
import numpy as np
import glob
from quica.measures.irr import *
from quica.dataset.dataset import IRRDataset
from quica.quica import Quica
pd.set_option('display.float_format', str)
np.random.seed(0)
pd.set_option('max_colwidth', 400)

import math

def trunc(num, digits):
    sp = str(num).split('.')
    return float('.'.join([sp[0], sp[1][:digits]]))
```

## Results 1

```python
import pandas as pd
import io 

cs_results_1 = """
0:0,0.27,0.33,0.58,0.66
2:2,0.33,0.28,0.56,0.57
4:4,0.11,0.36,0.44,0.64
4:2/2:4,0.44,0.44,0.55,0.65
4:0/0:4,0.45,0.55,0.68,0.70
2:0/0:2,0.41,0.60,0.55,0.68
"""

df = pd.read_csv(io.StringIO(cs_results_1), sep=",", names = ['1', '2', '3', '4'])
df['avg_acc'] = df.mean(axis=1)
df['avg_acc'] = df['avg_acc'].apply(lambda x: trunc(x, 2))
df['bias'] = ['without', 'without', 'without', 'with', 'with', 'with']
df[['bias','1','2','3','4', 'avg_acc']]
```

```python
np.mean(df['avg_acc'].tolist())
```

```python
df['1_and_2_'] = df[['1','2']].mean(axis=1)
df['1_and_2_'] = df['1_and_2_'].apply(lambda x: trunc(x, 2))

df
```

```python
without_section_1_and_2 = trunc(np.mean(df.query("bias == 'without'")[['1', '2']].mean(axis=1).tolist()),2)
# without_section_1_and_2
without_section_3 = np.mean(df.query("bias == 'without'")[['3']].mean(axis=1).values.tolist())
# without_section_3
without_section_4 = trunc(np.mean(df.query("bias == 'without'")[['4']].mean(axis=1).values.tolist()), 2)
# without_section_4
with_section_1_and_2 = trunc(np.mean(df.query("bias == 'with'")[['1', '2']].mean(axis=1).values.tolist()),2)
# with_section_1_and_2
with_section_3= trunc(np.mean(df.query("bias == 'with'")[['3']].mean(axis=1).values.tolist()),2)
# with_section_3
with_section_4 = trunc(np.mean(df.query("bias == 'with'")[['4']].mean(axis=1).values.tolist()),2)
# with_section_4
```

```python
cs_results_1 = f"""
1and2,{without_section_1_and_2},{with_section_1_and_2}
3,{without_section_3},{with_section_3}
4,{without_section_4},{with_section_4}
"""

df = pd.read_csv(io.StringIO(cs_results_1), names = ['without', 'with'],sep=",")
df['with_without'] = df.mean(axis=1).values.tolist()
df['with_without'] = df['with_without'].apply(lambda x: trunc(x,2))

df['delta_with'] = df['with'].diff()
df['delta_without'] = df['without'].diff()
# df['delta_with'] = df['delta_with'].apply(lambda x: trunc(x,2))
df['delta_with_without'] = df['with_without'].diff()
df['delta_with_without'] = df['delta_with_without'].apply(lambda x : round(x,2))
df['delta_with'] = df['delta_with'].apply(lambda x : round(x,2))
df['delta_without'] = df['delta_without'].apply(lambda x : round(x,2))

df['without'] = df['without'].apply(lambda x : round(x,2))



df[['without', 'delta_without', 'with', 'delta_with', 'with_without', 'delta_with_without']]
```
