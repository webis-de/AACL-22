```python
import os
import pandas as pd
import numpy as np
import glob



def get_labels_fix(e):
    if e == '1':
        return 'pro'
    elif e == '2':
        return 'con'
    elif e == '3':
        return 'unk'
    elif e == '4':
        return 'neutral'
    
    
def wrangle_argument(a_df, worker_id_col):
    hit = pd.concat([worker_id_col, a_df], axis=1)
    a_df = hit.set_index('id_worker')
    a_df = a_df.dot(a_df.columns).to_frame('original_answer').reset_index()
    a_df['original_answer'] = a_df['original_answer'].apply(lambda x : ' '.join(x.split('.')[-1:]))
    a_df['correct_ans'] = a_df['original_answer'].apply(lambda x : ' '.join(x.split('-')[1]))
    a_df['original_answer'] = a_df['original_answer'].apply(lambda x : ' '.join(x.split('-')[0]))
    a_df['id_worker'] = a_df['id_worker'].apply(lambda x : x[0])
    a_df['original_answer'] = a_df['original_answer'].apply(get_labels_fix)
    a_df['correct_ans'] = a_df['correct_ans'].apply(get_labels_fix)
    return a_df    
    
    
df = pd.concat(map(pd.read_csv, glob.glob(os.path.join('',
                                                       "/differential_bias/data/crowdsourced/raw/*.csv"))))
```

```python
hit_1 = pd.read_csv("../hit_1.csv").reset_index(drop=True)

worker_id_col_1 = hit_1[['id_worker']]
hit_1 = hit_1.iloc[:, 43:75].reset_index(drop=True)

arg_1 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 0:4]], axis = 1), worker_id_col_1)
arg_2 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 4:8]], axis = 1), worker_id_col_1)
arg_3 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 8:12]], axis = 1), worker_id_col_1)
arg_4 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 12:16]], axis = 1), worker_id_col_1)
arg_5 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 16:20]], axis = 1), worker_id_col_1)
arg_6 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 20:24]], axis = 1), worker_id_col_1)
arg_7 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 24:28]], axis = 1), worker_id_col_1)
arg_8 = wrangle_argument(pd.concat([worker_id_col_1, hit_1.iloc[:, 28:32]], axis = 1), worker_id_col_1)




from sklearn.metrics import accuracy_score, classification_report
hit_1 = pd.concat([arg_1, arg_2, arg_3, arg_4, arg_5, arg_6, arg_7, arg_8], axis = 0)

print('acc: ', accuracy_score(hit_1['original_answer'].tolist(), hit_1['correct_ans'].tolist()))

print(classification_report(hit_1['original_answer'].tolist(), hit_1['correct_ans'].tolist(), zero_division=0))
```

## Precision and accuracy

```python
data = pd.read_csv('../data/crowdsourced/prec_accs.csv', sep="\t")
data

distros = list(set(data['distro'].tolist()))
print("Dist|   prec avg  | acc avg  ")
print("---------------------------")
for e in distros:
    pro_vals = np.mean(data[data['distro']==e]['pro'].tolist())
    con_vals = np.mean(data[data['distro']==e]['con'].tolist())    
    acc_vals = np.mean(data[data['distro']==e]['acc'].tolist())
    print(e, '|', round(pro_vals,2), '|', round(con_vals,2), '|', round(acc_vals,2))
```
