```python
import pandas as pd
import numpy as np

def fix_col(e):
    if e in ['0:0', '2:2', '4:4']:
        return e
    else:
        if e in ['0:2', '2:0']:
            return '0:2/2:0'
        elif e in ['0:4', '4:0']:
            return '0:4/4:0'
        elif e in ['2:4', '4:2']:
            return '2:4/4:2'

        
    
df = pd.read_csv("../data/crowdsourced/confidence.csv", sep=',')

df
```

```python
with_bias = ['0:0', '2:2', '4:4']
without_bias = ['2:4/4:2', '0:4/4:0', '0:2/2:0']


print('          very_sure', '   reasonable_sure', '    unsure')
cs1 = df.query("setup==1")
for e in with_bias:
    very_sure = np.round(np.mean(cs1.query(f"distro=='{e}'")['very_sure'].tolist()),2)
    reasonable_sure = np.round(np.mean(cs1.query(f"distro=='{e}'")['reasonable_sure'].tolist()),2)
    unsure = np.round(np.mean(cs1.query(f"distro=='{e}'")['unsure'].tolist()),2)
        
    print(e, '       ',very_sure,'        ', reasonable_sure, '      ', unsure)
    
    
for e in without_bias:
    very_sure = np.round(np.mean(cs1.query(f"distro=='{e}'")['very_sure'].tolist()),2)
    reasonable_sure = np.round(np.mean(cs1.query(f"distro=='{e}'")['reasonable_sure'].tolist()),2)
    unsure = np.round(np.mean(cs1.query(f"distro=='{e}'")['unsure'].tolist()), 2)
        
    print(e, '   ',very_sure,'         ', reasonable_sure, '    ', unsure)

    
    
```
