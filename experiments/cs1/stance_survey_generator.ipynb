```python
from template_utils import *

root = path.abspath(path.join(os.getcwd(),"../"))
summary = read_json(root+'/data/processed/frame_stance_summary.json')
df = pd.read_csv(root+'/data/raw/Webis-argument-framing.csv')
df['argument'] = df[['premise', 'conclusion']].agg(' '.join, axis=1)
df = df[['argument_id', 'frame_id', 'frame', 'topic_id', 'topic', 'argument','stance']]


df['frame'] = df['frame'].apply(lambda x: x.strip())
df['topic'] = df['topic'].apply(lambda x: x.strip())
df['argument'] = df['argument'].apply(lambda x: x.strip())

df['num_words'] = df['argument'].apply(lambda x: len(x))
df['num_sentences'] = df['argument'].apply(get_arg_length)

df = df[df.num_sentences.between(1,4)]

print(len(summary))
```

```python

filtered = []
for e in summary:
    a_dict = {}
    extra_dict = {}
    for item, value in e['frame_info'].items():
        if isinstance(value, dict):
            pro = value['stance'].get('Con',0) >= 4 and value['stance'].get('Pro',0) >= 4
            if pro:
                a_dict[item] = value
        else:
            extra_dict[item] = value
    if a_dict:
        a_dict.update(extra_dict)
        b_dict = dict(e)
        b_dict['frame_info'] = a_dict
        filtered.append(b_dict)

filtered
```

```python
topics = []
frames = []
raw_questionnaire_list = []


for e in filtered:
    frames.append(list(list(e.values())[2].keys())[0])
    topics.append(e['topic_id'])

for x, y in zip(topics, frames):
    print('topic_id == {} and frame =="{}"'.format(x,y))
    temp = df.query('topic_id == {} and frame =="{}"'.format(int(x),y))
    raw_questionnaire_list.append(temp.reset_index())
    

```

```python
raw_questionnaire_list = list(filter(lambda x: not x.empty, raw_questionnaire_list))


raw_questionnaire_list_2 = []

for e in raw_questionnaire_list:
    if e.shape[0] < 8:
        pass
    else:
        raw_questionnaire_list_2.append(e)

for i,j in enumerate(raw_questionnaire_list_2):
    print(i, j['topic'].tolist()[0])
```
